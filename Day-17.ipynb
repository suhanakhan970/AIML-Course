Bar Plots

```python
name=['suhana','neha','depika','kirti','depali']
marks=[12,14,43,54,34,]
import matplotlib.pyplot as plt
plt.bar(name,marks,color='red',width=0.8)
plt.xlabel('name')
plt.ylabel('marks')
plt.title('marks')
```

Scatter plot

```python
x=['neha','isha','priya','kirti','suhana','naina','sania']
y=[20,30,45,65,67,89,88]
z=[54,56,76,34,34,32,22]
plt.scatter(x,y,color='green',marker='s')
plt.scatter(x,z,color='black',marker='o')
plt.xlabel('Names')
plt.ylabel('Marks')
plt.title('Marks')
```

Line Plot

```python
x=[2,3,5,4,8,9,7]
y=[8,8,6,5,4,9,7]
plt.plot(x,y,label='nvidia',color='red')
```

```python
x = [1,2,3,4,5] # years
a = [10,30,60,20,50] # sales
b = [5,10,50,20,15] # sales
c = [100,50,80,400,400] # sales
plt.plot(x,a,label='idea',color='black')
plt.plot(x,b,label='jio',color='pink')
plt.plot(x,c,label='airtel',color='yellow')
plt.xlabel('years')
plt.ylabel('sales')
plt.title('Company Sales Comparison')
plt.legend()
plt.show()
```
