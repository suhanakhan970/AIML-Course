```python
import matplotlib.pyplot as plt
import numpy as np
#sine wave parameter
x = np.linspace(0, 10, 100)

y=np.sin(x)
y2=np.sin(x + np.pi / 4)
y3=np.cos(x)
#y4=np.tan(x)
plt.plot(x, y, label='Sine Wave', color='blue')
plt.plot(x, y2, label='sine wave+pi/4',color='orange')
plt.plot(x,y3,label="cosine wave",color="yellow")
plt.title("sine,cosine,tangent wave")
plt.xlabel("time")
plt.ylabel("amplitude")
plt.legend()
plt.grid()
plt.show()
```

```python
def plot_sine_wave(frequency,duration,sample_rate):
    t = np.linspace(0, duration, int(sample_rate * duration), endpoint=False)
    y=np.sin(2*np.pi*frequency*t)
    plt.figure(figsize=(10,4))
    plt.plot(t,y)
    plt.title(f'Sine Wave: {frequency} Hz')
    plt.xlabel('time[s]')
    plt.ylabel('amplitude')
    plt.grid()                   
    plt.show()

    
```

```python
plot_sine_wave(frequency=5, duration=2, sample_rate=1000)
```
